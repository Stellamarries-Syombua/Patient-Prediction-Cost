# Patient Prediction Cost

## Introduction
This project aims to analyze and understand the factors influencing medical insurance charges using data obtained from Kaggle.
The dataset contains demographic and lifestyle characteristics such as age, BMI, smoking status and region.

## Objective
To explore the data and identify key patterns snd relationships influencing medical costs.


```python
# Load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted')  # modern equivalent
```

```python
# Load dataset
df = pd.read_csv("C:\\Users\\PC\\PycharmProjects\\PythonProject1\\data\\raw data\\insurance.csv")
df.head() # Check the first few rows
```

## Data Overview
Understanding the structure and composition of data


```python
df.info() # Summary of the dataframe

df.describe() # Generate descriptive stats

df.isnull().sum()
```

## Variable Types
- Numerical : age, bmi, children, charges
- Categorical : sex. smoker, region
- No missing values

# Exploratory Data Analysis
## Distribution of Medical Charges
Analyzing how medical charges are distributed among individuals

```python
plt.figure()
sns.histplot(df['charges'], kde=True)
plt.title('Medical Charges distribution')
plt.xlabel('Charges')
plt.ylabel('Frequency')
plt.show()
```

## Smoking status vs charges
Investigating the impact of smoking on medical charges


```python
plt.figure()
sns.boxplot(x = 'smoker', y = 'charges', data = df)
plt.title('Charges by Smoking Status')
plt.show()
```

## BMI Vs Charges
Exploring the relationship between bmi and medical charges


```python
plt.figure()
sns.scatterplot(x = 'bmi', y = 'charges', data =df)
plt.title('BMI vs Charges')
plt.show()
```

## Age vs Charges
Understanding how age influences medical charges

```python
plt.figure()
sns.scatterplot(x = 'age', y = 'charges', data = df)
plt.title('Age vs Charges')
plt.show()
```

## Children vs Charges
Examining whether the number of the dependants affects medical costs

```python
plt.figure()
sns.boxplot(x = 'children', y = 'charges', data = df)
plt.title('Children vs Charges')
plt.show()
```
